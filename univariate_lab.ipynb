```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

%matplotlib inline
```

```python
file_path="/Users/home/Desktop/Rao_labs/lab-eda-univariate/amz_uk_price_prediction_dataset.csv"
df = pd.read_csv(file_path)
```

```python
frequency_table = df['category'].value_counts()
frequency_table
```

```python
top5=frequency_table.sort_values(ascending=False).head(5)
plt.figure(figsize=(10, 5))
sns.barplot(x=top5.index, y=top5.values, hue=top5.index,palette="Set3")
plt.xticks(rotation=45)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Prepare data
labels = top5.index
sizes = top5.values
colors = sns.color_palette("Set3")

# Create figure and pie chart with label distance adjusted
plt.figure(figsize=(8, 15))
plt.pie(
    sizes,
    labels=labels,
    autopct='%1.1f%%',
    startangle=90,
    colors=colors,
    labeldistance=1.15  # 👈 move labels away from the center
)

plt.title("Top 5 Categories by Sales")
plt.tight_layout()
plt.show()
```

```python

mean_price=df['price'].mean()
median_price=df['price'].median()
mode_price=df['price'].mode()[0]

print(f"The mean price is {mean_price}")
print(f"The median price is {median_price}")
print(f"The mode price is {mode_price}")
#the average price is way above the mode price
```

```python
#Determine the variance, standard deviation, range, and interquartile range for product `price`.
variance_price = df['price'].var()
std_dev_price = df['price'].std()
min_price = df['price'].min()
max_price = df['price'].max()
range_price = max_price - min_price
quantiles_price = df['price'].quantile([0.25, 0.5, 0.75])

print(f"The price variance is {variance_price}")
print(f"The standard deviation price is {std_dev_price}")
print(f"The price range is {range_price}") 
print(f"The quantile price is {quantiles_price}")

#The price varies greatly
```

```python
#plt.figure(figsize=(10,6))
sns.histplot(df['price'],bins=5,color='salmon')
plt.title('Distribution of Product Prices')
plt.xlabel('price')
plt.ylabel('Frequency')
#plt.grid(True)
#plt.tight_layout()
plt.show()
```

```python
#Are there products that are priced significantly higher than the rest? Use a box plot to showcase the spread and potential outliers in product pricing. 
sns.boxplot(data = df['price'], color="lightblue")
plt.show()
```

```python
# Calculate the mean, median, and mode for the `rating` of products.
mean_rating=df['stars'].mean()
median_rating=df['stars'].median()
mode_rating=df['stars'].mode()

print(f"The mean rating is {mean_rating}")
print(f"The median rating is {median_rating}")
print(f"The mode rating is {mode_rating}")
```

```python
#Determine the variance, standard deviation, and interquartile range for product `rating`.
var_rating=df['stars'].var()
std_rating=df['stars'].std()
quantiles_rating=df['stars'].quantile([0.25, 0.5, 0.75])

print(f"The variance of rating is {var_rating}")
print(f"The standard deviation of rating is {std_rating}")
print(f"The quantiles rating is {quantiles_rating}")
```

```python
#Calculate the skewness and kurtosis for the `rating` column.
skewness_rating = df['stars'].skew()
kurtosis_rating = df['stars'].kurtosis()

print(f"The skewness of rating is {skewness_rating}")
print(f"The kurtosis rating is {kurtosis_rating}")
```
